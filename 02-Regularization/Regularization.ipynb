## Regularization

```python
import pandas as pd
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
correlations = train.corr()
correlations
```

```python
saleprice_correlations = correlations["SalePrice"]
saleprice_correlations
```

```python
X_train = train.drop('SalePrice', axis=1)
y_train = train.loc[:, 'SalePrice']
```

```python
X_test = test.drop('SalePrice', axis=1)
y_test = test.loc[:, 'SalePrice']
```

```python
from sklearn.linear_model import LinearRegression, Ridge
```

```python
linear_reg = LinearRegression()
ridge_reg = Ridge(alpha=0.05, normalize=True)
```

```python
linear_reg.fit(X_train, y_train)
ridge_reg.fit(X_train, y_train)
```

```python
linear_pred = linear_reg.predict(X_test)
ridge_pred = ridge_reg.predict(X_test)
```

```python
comparison_linear_reg = pd.DataFrame({"Actual Values" : y_test, "Predictions" : linear_pred})
```

```python
comparison_linear_reg.head()
```

```python
comparison_ridge_reg = pd.DataFrame({"Actual Values" : y_test, "Predictions" : ridge_pred})
```

```python
comparison_ridge_reg.head()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
linear_mse = mean_squared_error(y_test, linear_pred)
ridge_mse = mean_squared_error(y_test, ridge_pred)
```

```python
print(f'MSE without Ridge: {linear_mse}')
print(f'MSE with Ridge: {ridge_mse}')
```
